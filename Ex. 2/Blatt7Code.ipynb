```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['font.family'] = 'STIXGeneral'

R = 1 # m
Q = 1 # C#
def Elow(r):
    return (1/(4*np.pi)) * Q*r / R**3
    
def Ehigh(r):
    return 1/(4*np.pi) * Q / r**2

x = np.linspace(0, 1, 1000)

plt.plot(x, Elow(x), color = 'blue')

x = np.linspace(1, 5, 1000)
plt.plot(x, Ehigh(x), color = 'blue')

plt.xlabel(r'$r \: [\text{m}]$')
plt.ylabel(r'$Q \: \left[ \frac{\text{C}}{\epsilon _0} \right]$')
plt.grid()
plt.xlim([0, 5])
plt.ylim([0, 0.1])
plt.title(r'$\left| \vec E(\vec r) \right|$ für $Q = 1 \text{C}; \: R=1 \text{m}$')
plt.savefig('Ladungsverteilung-Vollkugel.png', dpi=1200)
```
