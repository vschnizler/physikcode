```python
import sympy as sym
from sympy import *
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['font.family'] = 'STIXGeneral'


def E(x):
    return (8.854e-12)*((-3e-6/(x**2)) + (7e-6/((x-0.4)**2)))

x = np.linspace(-0.79, -0.01, 1000)

plt.plot(x, E(x))
plt.ylabel(r'$ \vec E (x) $', fontsize=12)
plt.xlabel(r'$\frac{x}{m}$', fontsize=20)
plt.grid()
plt.title(r'Ladungsverteilung f√ºr $x < 0$', fontsize=15)
plt.tight_layout()
plt.show()

x, y, z = sym.symbols('x y z')

v1 = 2*x*y + 2*x*z**2 + 3*x**2
v2 = x**2 + z**2 + 2*y
v3 = 2*y*z + 2*x**2 * z + 1

C = sym.coordo('C')
v = sym.matrix_to_vector(sym.ImmutableDenseMatrix([v1, v2, v3]), C)
sym.curl(v)
```
